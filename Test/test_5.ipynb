```python
import csv
import matplotlib.pyplot as plt

# Tạo và ghi dữ liệu vào file CSV
filename = "data.csv"
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["X", "Y"])  # Ghi tiêu đề cột
    for i in range(101):  # X chạy từ 0 đến 10
        y = i ** 2  # Hàm số y = x^2
        writer.writerow([i, y])

# Đọc dữ liệu từ file CSV
x_values = []
y_values = []
with open(filename, mode='r') as file:
    reader = csv.reader(file)
    next(reader)  # Bỏ qua hàng tiêu đề
    for row in reader:
        x_values.append(int(row[0]))
        y_values.append(int(row[1]))

# Vẽ đồ thị
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='y = x^2')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Graph of y = x^2")
plt.legend()
plt.grid()
plt.show()
```
