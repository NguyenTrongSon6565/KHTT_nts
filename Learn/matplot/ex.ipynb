```python
# vẽ đồ thị các điểm biểu diễn chuyển động của một vật được ném xiên
import matplotlib.pyplot as plt
import numpy as np
g = 9.8
v0 = 10
theta = np.pi/4
t = np.linspace(0, 2*v0*np.sin(theta)/g, 50)
x = v0*np.cos(theta)*t
y = v0*np.sin(theta)*t - 0.5*g*t**2
plt.scatter(x, y, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Motion of a projectile')
plt.legend(['Projectile'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def DoThiNemXien(v0, theta_deg, g=9.8):
    """
    Vẽ đồ thị quỹ đạo của chuyển động ném xiên.
    
    Tham số:
    - v0: Vận tốc ban đầu (m/s)
    - theta_deg: Góc ném (độ)
    - g: Gia tốc trọng trường (m/s^2), mặc định là 9.8
    
    Trả về:
    - t: Mảng thời gian
    - x, y: Mảng tọa độ theo các thời điểm của t
    """
    theta = np.radians(theta_deg)  # Chuyển đổi góc sang radian
    
    # Thời gian bay tối đa
    t_max = 2 * v0 * np.sin(theta) / g
    
    # Chia nhỏ thời gian để vẽ đồ thị
    t = np.linspace(0, t_max, num=20)
    
    # Tính tọa độ x, y theo thời gian
    x = v0 * np.cos(theta) * t
    y = v0 * np.sin(theta) * t - 0.5 * g * t**2
    
    # Lưu tọa độ vào danh sách
    toa_do = list(zip(t, x, y))
    
    # In ra tọa độ theo thời gian
    for time, x_val, y_val in toa_do:
        print(f"t={time:.2f}s, x={x_val:.2f}m, y={y_val:.2f}m")
    
    # Vẽ đồ thị quỹ đạo
    plt.figure(figsize=(8, 5))
    plt.scatter(x, y, label=f"Quỹ đạo ném xiên (v0={v0} m/s, θ={theta_deg}°)")
    plt.xlabel("Tọa độ x (m)")
    plt.ylabel("Tọa độ y (m)")
    plt.title("Đồ thị quỹ đạo ném xiên")
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.legend()
    plt.grid()
    plt.show()
    
    return t, x, y

# Ví dụ sử dụng
t_vals, x_coords, y_coords = DoThiNemXien(10, 30)
```
