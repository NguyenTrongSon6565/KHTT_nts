# SciPy là một hàm chưa các thông thức toán học phức tạp.

```python
from scipy import integrate # integrate là hàm tính tích phân
from math import sin
from math import exp
from math import pi
import numpy as np
import matplotlib.pyplot as plt

def myfunction(x):
    # you need to define result
    result = x**2
    return result

integral = integrate.quad(myfunction, 0, 4) # quad là hàm cho phép tính chạy từ a đến b
print(integral)

def function_e(x):
    result = exp(x)
    return result

integral_e = integrate.quad(function_e, 0, pi) # tính từ 0 đến pi
print(integral_e)
```

```python
# tính tích phân của hàm sin(x)d.dx từ 0 đến 1
def function_sin(x):
    result = sin(x)
    return result

integral_sin = integrate.quad(function_sin, 0, 1)
print(integral_sin) 
```

```python
# tạp hàm vẽ đồ thị chung
def plot_function(func, a, b, label):
    x = np.linspace(a, b, 100)
    y = func(x)
    plt.plot(x, y, label=label)
    plt.legend()
    plt.show()
```

```python
# ve do thi cua ham sin(x)
plot_function(np.sin, 0, 20, 'sin(x)')
plt.savefig('sinx.png')
```

```python
# create function simpson's rule
def simpson_rule(func, a, b, n):
    h = 1.0*(b - a) / n # n even interger
    sum1 = 0
    sum2 = 0
    for i in range(1, (n//2) + 1):
        sum1 += func(a + (2 * i - 1) / h)
    for i in range(1, (n//2)):
        sum2 += func(a + 2 * i * h)

    result = (h / 3) * (func(a) + 4 * sum1 + 2 * sum2 + func(b))
    return result

def func(x):
    return 3 * (sin(x)**3) / 2

print(simpson_rule(func, 0, pi, 500)) 

```
